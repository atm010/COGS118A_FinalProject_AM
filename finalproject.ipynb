```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# data sets that I will use from UCI ML Repository
# datasets are IRIS, BREAST_CANCER, and DIGITS

dataset1 = datasets.load_iris()
dataset2 = datasets.load_breast_cancer()
dataset3 = datasets.load_digits()
datasetslist = [dataset1, dataset2, dataset3]

# declaring how many trials there will be

trials = 3

# classifiers
# the three classifiers are SVM, KNN, and Logistic Regression

classifier0 = SVC(kernel='linear', random_state=0)
classifier1 = KNeighborsClassifier(n_neighbors = 5, metric='minkowski',p=2)
classifier2 = LogisticRegression(random_state=0)
classifiers = []
classifiers.append(classifier0)
classifiers.append(classifier1)
classifiers.append(classifier2)
classifieraccuracydict = {0:[],1:[],2:[]}

# for loop to iterate through datasets

for i in range(len(datasetslist)):
    X = datasetslist[i].data
    Y = datasetslist[i].target
    print("Dataset"+str(i))
    print("Partitions: 0.2,0.5,0.8")
    print("Trials: "+str(trials))
    
    # for loop to iterate through partitions
    # partitions of 20/80, 50/50, 80/20
    
    for j in [0.2,0.5,0.8]:
        X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = j,random_state=0)
        
        #for loop to iterate through trials
        
        for t in range(trials):
            
            #for loop to iterate through classifiers
            
            for c in range(len(classifiers)):
                print("Classifier"+str(c))
                classifiers[c].fit(X_train, Y_train)
                accuracies = cross_val_score(estimator=classifiers[c], X=X_train,y=Y_train,cv=10)
                print(accuracies.mean())
                classifieraccuracydict[c].append(accuracies.mean())
                              
```

```python
print("CLassifier vs Accuracy")
print("Dataset1" + str(classifieraccuracydict[0]))
print("Dataset2" + str(classifieraccuracydict[1]))
print("Dataset3" + str(classifieraccuracydict[2]))
print("Final classifier accuracy")
print("SVM: "+str(sum(classifieraccuracydict[0])/len(classifieraccuracydict[0])))
print("KNN: "+str(sum(classifieraccuracydict[1])/len(classifieraccuracydict[1])))
print("Logistic: "+str(sum(classifieraccuracydict[2])/len(classifieraccuracydict[2])))  
```
